```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

from astropy import constants as cst
from astropy import units as u
import healpy as hp

from aqn import *
from constants import *
from survey_parameters import *
from skymap_plotting_functions import *
from aqn_simulation import *
```

```python


h_func_cutoff = 17 + 12*np.log(2)
def h(x):
    return_array = np.copy(x)
    return_array[np.where(x<1)] = (17 - 12*np.log(x[np.where(x<1)]/2))
    return_array[np.where(x>=1)] = h_func_cutoff
    return return_array

# erg Hz^-1 s^-1 cm^-2
# nu Hz
# T eV
def spectral_surface_emissivity(nu, T):
    T = T * eV_to_erg
    w = 2 * np.pi * nu * Hz_to_erg
    unit_factor = (1 / cst.hbar.cgs) * (1/(cst.hbar.cgs * cst.c.cgs))**2 * (cst.hbar.cgs * 1/u.Hz * 1/u.s)
    #                ^ 1/seconds           ^ 1/area                          ^ 1/frequency and energy
    return unit_factor * 4/45 * T**3 * cst.alpha ** (5/2) * 1/np.pi * (T/(m_e_eV*eV_to_erg))**(1/4) * (1 + w/T) * np.exp(- w/T) * h(w/T)

def convert_to_skymap_units(F_erg_hz_cm2,nu):
    erg_hz_cm2 = 1*u.erg/u.s/u.Hz/u.cm**2
    C = erg_hz_cm2.to(skymap_units)
    return (F_erg_hz_cm2 * C / erg_hz_cm2).to(u.mK, equivalencies = u.brightness_temperature(nu))

########### ionized gas #############################################
def rho_ioni_gas(x, parameters):
    a1, b1, c1 = parameters[0]*1e-23, parameters[1]*10, parameters[2] 
    a2, b2, c2 = parameters[3]*1e-23, parameters[4], parameters[5]
    a3         = parameters[6]*1e-24

    # a1, b1, c1 = 6.2e-23, 1.2, 10
    # a2, b2, c2 = 1.3e-23, 1.3, 1
    # a3         = 0.15e-23
    
    ret = a1 / (1+(x/b1)**2)**c1 + \
          a2 / (1+(x/b2)**2)**c2 + \
          a3
    
    return ret
#####################################################################

########### neutral gas #############################################
def rho_neut_gas(x, parameters):
    a1, b1, c1 = parameters[0]*1e-21, parameters[1], parameters[2] 
    a2, b2, c2 = parameters[3]*1e-21, parameters[4], parameters[5]
    a3         = parameters[6]*1e-24

    ret = a1 / (1+(x/b1)**2)**c1 + \
          a2 / (1+(x/b2)**2)**c2 + \
          a3
    
    return ret
#####################################################################

########### dv ionized #############################################
def rho_dv_ioni(x, parameters):
    a1, b1, c1 = parameters[0]*1e5, parameters[1]*10, parameters[2]*10 
    a2, b2, c2 = parameters[3]*1e5, parameters[4], parameters[5]
    a3         = parameters[6]*1e5
    
    ret = a1 / (1+(x/b1)**2)**c1 + \
          a2 / (1+(x/b2)**2)**c2 + \
          a3
    
    return ret
#####################################################################

########### dv neutral #############################################
def rho_dv_neut(x, parameters):
    a1, b1, c1 = parameters[0]*1e5, parameters[1]*1e2, parameters[2]*1e2 
    a2         = parameters[3]*1e5
    
    ret = a1 / (1+(x/b1)**2)**c1 + \
          a2
    
    return ret
#####################################################################

########### ionized gas temperature ################################
def rho_temp(x, parameters):
    a1, b1, c1 = parameters[0]*1e7, parameters[1], parameters[2]
    a2, d2     = parameters[3]*1e7, parameters[4]
    a3         = parameters[5]*1e7
    
    ret = a1 / (1+(x/b1)**2)**c1 + \
          a2 * np.tanh(x-d2) + a3
    
    return ret
#####################################################################


def rho_dark_mat(R, parameters):
    rho_s, r_s, alpha, beta, gamma = parameters
        
    R_kpc   = R  * u.kpc
    rho_s_msol_kpc3 = rho_s * u.solMass / u.kpc**3 * 1e7
    r_s_kpc = r_s * u.kpc
    
    rho_s_kg_m3 = (rho_s_msol_kpc3).to(u.kg/u.m**3)
    
    return (( (2**((beta - gamma)/alpha)) * rho_s_kg_m3 ) / ( (R_kpc/r_s_kpc)**gamma * (1+(R_kpc/r_s_kpc)**alpha)**((beta - gamma)/alpha)  )).value
```

```python
# define AQN parameters
m_aqn_kg = 0.05*u.kg
R_aqn_cm = calc_R_AQN(m_aqn_kg)

print("using m_aqn =", m_aqn_kg)
print("using R_aqn =", R_aqn_cm)
```

```python
# specify frequency resolution and create frequency band array
frequency_band_resolution = 1e9
frequency_band = np.arange(f_min_hz.value, f_max_hz.value, frequency_band_resolution) * u.Hz
```

```python
# cube_indexes = np.load("../data/FIRE/skymap_integration_cube_indexes.npy")
sun_mw_distance_kpc = 8*u.kpc
NSIDE = 2**5
NPIX = hp.nside2npix(NSIDE)
dOmega = hp.nside2pixarea(nside=NSIDE)

# array of theta, phi, for each pixel
theta, phi = hp.pix2ang(nside = NSIDE, ipix = list(range(NPIX)))
dl =  1 * u.kpc
l_min, l_max = 0.01, 10
l_list = np.arange(l_min,l_max,dl.value)[:, np.newaxis]
# 2D array of distance elements along each sightline
l = np.ones((len(l_list),len(theta))) * l_list

# (l,theta,phi) -> (x,y,z)
x = l * np.sin(theta) * np.cos(phi) - sun_mw_distance_kpc.value
y = l * np.sin(theta) * np.sin(phi)
z = l * np.cos(theta)
R = np.sqrt(x**2 + y**2 + z**2)
```

```python
ioni_gas = (rho_ioni_gas(R, [6.521713155920115 , 0.2940077939925656, 
                                    29.072973590609415, 1.6812716453492722, 
                                    3.9044895415071528, 3.4656473013927394, 
                                    1.43427732961435])*u.kg/u.m**3/cst.m_p.si).to(1/u.cm**3) 
neut_gas = (rho_neut_gas(R, [8.97656111186945  ,2.213270305404448 ,
                                    13.875013340031673,1.5018739204854163,
                                    2.58047607932149  ,1.1680782169590678,29.693878727525025])*u.kg/u.m**3/cst.m_p.si).to(1/u.cm**3)
dark_mat = (rho_dark_mat(R, [19.60758081904341, 1.081829387522867, 
                                    2.207828893849368, 1.831142772499595, 
                                    0.06711443461154958])*u.kg/u.m**3/  m_aqn_kg).to(1/u.cm**3)
temp_ion = rho_temp(R, [6.14, 0.3, 1.5, 0.1, 3.3, 0.1]) * u.K * K_to_eV
dv_ioni  = rho_dv_ioni(R, [0.7594266933349966, 2.73608605904013, 
                                     2.5582050186794847, 1.004084602553337, 
                                     1.6631635389592465, 17.152525652652074, 
                                     2.7163675354935055]) * u.m/u.s /cst.c
dv_neut  = rho_dv_neut(R, [0.7817275122369671, 1.4514477940369448, 
                                     7.670255389377027 , 
                                     2.7444837892336853]) * u.m/u.s /cst.c
```

```python
# compute T_AQN
t_aqn_n = T_AQN_analytical(neut_gas, dv_neut, f, g)
t_aqn_i = T_AQN_ionized2(ioni_gas, dv_ioni, f, g, temp_ion, R_aqn_cm)
# if ioni temp is lower than neut temp, replace with neut temp
t_aqn_i[t_aqn_i<t_aqn_n] = t_aqn_n[t_aqn_i<t_aqn_n]
```

```python
aqn_emit = np.zeros(np.shape(t_aqn_n)) *  u.mK # (u.erg/(u.Hz*u.s*u.cm**2)) 

for nu in frequency_band[0:1]:
    aqn_emit += convert_to_skymap_units(spectral_surface_emissivity(nu, t_aqn_n)/(dOmega*u.sr), nu) + \
                convert_to_skymap_units(spectral_surface_emissivity(nu, t_aqn_i)/(dOmega*u.sr), nu)
    
aqn_emit = aqn_emit * 4 * np.pi * R_aqn_cm**2 * dark_mat
```

```python
aqn_emit
```

```python
f_aqn = np.sum(aqn_emit, axis=0) / (4*np.pi) * (dl.cgs) * dOmega
```
